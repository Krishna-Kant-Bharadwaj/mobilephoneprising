```python
# ==============================
# STEP 1: Install required packages
# ==============================
!pip install xgboost lightgbm scikit-learn seaborn matplotlib

# ==============================
# STEP 2: Import dependencies
# ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
import lightgbm as lgb

# ==============================
# STEP 3: Load the dataset
# ==============================
from google.colab import drive
drive.mount('/content/drive')
data = pd.read_csv("/content/drive/My Drive/mobile_phone_pricing/Mobile Phone Pricing/dataset.csv")
print(data.head())
print(data.info())

# ==============================
# STEP 4: Quick exploration
# ==============================
print("\nMissing values per column:\n", data.isna().sum())
print("\nTarget variable counts:\n", data['price_range'].value_counts())

sns.countplot(x="price_range", data=data)
plt.title("Price Range Distribution")
plt.show()

# ==============================
# STEP 5: Split features/target
# ==============================
X_features = data.drop(columns=['price_range'])
y_target = data['price_range']

scaler_obj = StandardScaler()
X_normalized = scaler_obj.fit_transform(X_features)

X_train_set, X_test_set, y_train_set, y_test_set = train_test_split(
    X_normalized, y_target, test_size=0.2, random_state=123, stratify=y_target
)

print("Training samples:", X_train_set.shape, "Testing samples:", X_test_set.shape)

# ==============================
# STEP 6: Logistic Regression (baseline)
# ==============================
lr_model = LogisticRegression(max_iter=600)
lr_model.fit(X_train_set, y_train_set)
lr_preds = lr_model.predict(X_test_set)
print("\nLogReg Accuracy:", accuracy_score(y_test_set, lr_preds))

# ==============================
# STEP 7: Random Forest Classifier
# ==============================
rf_model = RandomForestClassifier(n_estimators=250, random_state=123)
rf_model.fit(X_train_set, y_train_set)
rf_preds = rf_model.predict(X_test_set)
print("\nRandom Forest Accuracy:", accuracy_score(y_test_set, rf_preds))

# ==============================
# STEP 8: XGBoost Classifier
# ==============================
xgb_model = xgb.XGBClassifier(eval_metric="mlogloss", use_label_encoder=False, random_state=123)
xgb_model.fit(X_train_set, y_train_set)
xgb_preds = xgb_model.predict(X_test_set)
print("\nXGBoost Accuracy:", accuracy_score(y_test_set, xgb_preds))

# ==============================
# STEP 9: LightGBM Classifier
# ==============================
lgb_model = lgb.LGBMClassifier(random_state=123)
lgb_model.fit(X_train_set, y_train_set)
lgb_preds = lgb_model.predict(X_test_set)
print("\nLightGBM Accuracy:", accuracy_score(y_test_set, lgb_preds))

# ==============================
# STEP 10: Model Evaluation
# ==============================
final_predictions = lgb_preds   # choose the best performing modelâ€™s predictions

print("\nDetailed Report:\n", classification_report(y_test_set, final_predictions))

matrix = confusion_matrix(y_test_set, final_predictions)
plt.figure(figsize=(6,5))
sns.heatmap(matrix, annot=True, cmap="Greens", fmt="d",
            xticklabels=[0,1,2,3], yticklabels=[0,1,2,3])
plt.title("Confusion Matrix")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.show()

# ==============================
# STEP 11: Custom Inference
# ==============================
def classify_mobile(input_features: dict):
    """
    input_features: dictionary of feature_name:value pairs
    Example:
    {"ram": 2048, "battery_power": 1500, "px_height": 600, ...}
    """
    temp_df = pd.DataFrame([input_features])
    temp_scaled = scaler_obj.transform(temp_df)
    result = lgb_model.predict(temp_scaled)[0]
    return result

# Example test
sample_case = X_features.iloc[5].to_dict()
print("\nPredicted Price Category:", classify_mobile(sample_case))
```
